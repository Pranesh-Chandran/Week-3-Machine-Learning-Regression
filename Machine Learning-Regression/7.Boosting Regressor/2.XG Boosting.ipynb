```python
import pandas as pd
import xgboost
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size =0.30, random_state=0)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
#Model Creation
from xgboost import XGBRegressor
```

```python

regressor = XGBRegressor(n_estimators = 50, random_state=0,max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#Save model
#Index(['age', 'bmi', 'children', 'charges', 'sex_male', 'smoker_yes'], dtype='object')
import pickle
filename = "Finalized_model_XG_regressor.sav"
pickle.dump(regressor, open(filename, 'wb'))
age = float(input("Enter your age :"))
bmi = float(input("Enter your bmi :"))
children = int(input("Enter your children count :"))
sex = int(input("Enter your sex Male =1 or female =0 :"))
smoker = int(input("Enter smoker = 1 otherwise = 0  :"))
load_model = pickle.load(open(filename,'rb'))
```

```python
r_result = load_model.predict([[age,bmi,children,sex,smoker]])
r_result
```
