```python
import pandas as pd
import lightgbm
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset, dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size =0.30, random_state=0)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
#Model Creation
#criterion{“squared_error”, “absolute_error”, “friedman_mse”, “poisson”}, default=”squared_error”
from lightgbm import LGBMRegressor
```

```python

regressor = LGBMRegressor()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

```python
## Save model
#Index(['age', 'bmi', 'children', 'charges', 'sex_male', 'smoker_yes'], dtype='object')
import pickle
filename = "Finalized_model_LGBM_regressor.sav"
pickle.dump(regressor, open(filename, 'wb'))
age_input = int(input("Enter your age :"))
bmi_input = int(input("Enter your bmi :"))
children_input = int(input("Enter your children count :"))
sex_male_input = int(input("Enter your sex Male =1 or female =0 :"))
smoker_yes_input = int(input("Enter smoker = 1 otherwise = 0  :"))
load_model = pickle.load(open(filename,'rb'))
r_result = load_model.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
r_result
```
