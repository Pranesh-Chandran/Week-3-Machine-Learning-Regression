```python
import pandas as pd #pandas library contains set of functions, its used for read the table, files and database as well
```

```python
dataset = pd.read_csv("Salary_Data.csv") #Read the data from database and print the data set
dataset
```

```python
independent = dataset[["YearsExperience"]] # assign the columnto individual variable like 'X-axis' and 'Y-axis'
dependent = dataset[["Salary"]]
```

```python
independent,dependent # print the column data
```

```python
from sklearn.model_selection import train_test_split #sklearn is the library contains set of class and function
```

```python
train_test_split(independent,dependent,test_size = 0.30, random_state = 0)
x_train,y_train,x_test,y_test = train_test_split(independent,dependent,test_size = 0.30, random_state = 0)
```

```python
#from sklearn.linear_model import LinearRegression
```

```python
#regressor = LinearRegression()
#regressor
```

```python
from sklearn.linear_model import LinearRegression #Regression algorthim is called under linear_model
```

```python
regressor = LinearRegression() #regressor object is created
```

```python
regressor
```

```python
regressor.fit(x_train,y_train) #fit function is called to find weight and bias value
```

```python
weight = regressor.coef_
bias = regressor.intercept_
```

```python
weight
```

```python
bias
```

```python
y_pred = regressor.predict(x_test) #prediction value is find for xtest(30% of input)
from sklearn.metrics import r2_score #R2 square function imported from metric calss
```

```python
r_score = r2_score(y_test,y_pred) # compare actual value and predicted value
```

```python
r_score
```
